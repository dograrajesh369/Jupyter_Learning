## Style sheets reference

https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html  

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style

plt.style.use('bmh')

x = [25,35,78,98,85]
y = [20,30,90,50,75]
x1 = [22,54,32,45,85]
y1 = [15,25,45,85,70]
plt.plot(x , label = "A")
plt.plot(y, label = "B")
plt.plot(x1, label = "C")
plt.plot(y1, label = "D")
plt.title("RD Raw Data")
plt.ylabel("This is Y")
plt.xlabel("This is X")
plt.legend()
plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style

plt.style.use('bmh')

x = [25,35,78,98,85]
y = [20,30,90,50,75]
x1 = [22,54,32,45,85]
y1 = [15,25,45,85,70]
plt.scatter(x,y , label = "A")
plt.scatter(x1,y1, label = "B")

plt.title("RD Raw Data")
plt.ylabel("This is Y")
plt.xlabel("This is X")
plt.legend()
plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style

plt.style.use('bmh')

x = [25,35,78]
y = [20,30,90]
x1 = [22,54,32]
y1 = [15,25,45]
plt.hist(x,y , label = "A" ,linewidth= 4)
plt.hist(x1,y1, label = "B",linewidth= 4)

plt.title("RD Raw Data")
plt.ylabel("This is Y")
plt.xlabel("This is X")
plt.legend()
plt.show()
```
